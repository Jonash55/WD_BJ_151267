### Obsługa wartości odstających

```python
p0 = df.Pregnancies.min()
p100 = df.Pregnancies.max()
q1 = df.Pregnancies.quantile(0.25)
q2 = df.Pregnancies.quantile(0.5)
q3 = df.Pregnancies.quantile(0.75)
iqr = q3 - q1
# wynik
p0, p100, q1, q2, q3, iqr
```

```python
lc = q1 - 1.5 * iqr
uc = q3 + 1.5 * iqr
lc, uc
```

```python
df.Pregnancies.plot(kind='box')
```

```python
df.Pregnancies.clip(upper=uc, inplace=True)
```

```python
df.Pregnancies.plot(kind='box')
```
